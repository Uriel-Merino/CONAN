# Run CONAN from configfile

```python
import numpy as np
import CONAN3
```

- load configfile to generate the lightcurve, rv and setup objects.

```python
lc_data, rv_data, mcmc = CONAN3.load_configfile("input_config_w127_example.dat",verbose=False)
```

```python
lc_data
```

```python
lc_data.plot()
```

```python
result = CONAN3.fit_data(lc_data,rv_data,mcmc)
```

- alternatively the fit can be perfomed directly from the `load_configfile` function by setting `return_fit`= True. This then additional outputs the `result` object

lc_data, rv_data, mcmc, $\textbf{result}$ = CONAN3.load_configfile( "input_config_w127_example.dat", 
                                                            $\textbf{return_fit=True}$, verbose=False)
